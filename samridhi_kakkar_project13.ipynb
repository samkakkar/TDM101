## Question 4

```r
myDF <- fread("/anvil/projects/tdm/data/taxi/yellow/yellow_tripdata_2015-11.csv", tz="")
```

```r
thanksgivingdayDF <- subset(myDF, grepl("2015-11-26", tpep_pickup_datetime))
dim(thanksgivingdayDF)  
```

```r
paradestart <- strptime("2015-11-26 09:00:00", format="%Y-%m-%d  %H:%M:%S", tz="EST")
paradeend <- strptime("2015-11-26 12:00:00", format="%Y-%m-%d  %H:%M:%S", tz="EST")
```

```r
mytimes <- strptime(thanksgivingdayDF$tpep_pickup_datetime, format="%Y-%m-%d  %H:%M:%S", tz="")
```

```r
finalDF <- subset(thanksgivingdayDF, (mytimes > paradestart) & (mytimes < paradeend))
dim(finalDF) 
```

->loaded taxi cab data using fread
-> filtered the data for a given day
-> defined the parade timings
-> converted the pickup times to datetime format
-> filtered the rides during the parade time

## Question 5

```r
testDF <- data.frame( finalDF$pickup_latitude, finalDF$pickup_longitude)
names(testDF) <- c("lat","long")
```

```r
points <- st_as_sf(testDF, coords = c("long", "lat"), crs = 4326)
```

```r
addCircleMarkers(addTiles(leaflet(points)), radius = 0.1)
```
